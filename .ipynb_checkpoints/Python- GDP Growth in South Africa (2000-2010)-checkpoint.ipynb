# Analysis of GDP Growth in South Africa

# Introduction
# This notebook analyzes the GDP growth rate of South Africa from 2000 to 2010.
# We will load the dataset, clean the data, explore trends,
# and visualize the results to understand the economic performance during this period.

```python
#Data Loading Cell

import pandas as pd

# Load the dataset
data = pd.read_csv('table1.1.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Displaying basic information about the dataset
data_info = data.info()
data_description = data.describe()

data_info, data_description
```

```python
# Data Visualization Cell

import matplotlib.pyplot as plt

# Plot GDP % growth over the years
plt.figure(figsize=(10, 5))
plt.plot(data['Year'], data['GDP % growth'], marker='o')
plt.title('GDP % Growth in South Africa (2000-2010)')
plt.xlabel('Years')
plt.ylabel('GDP % Growth')
plt.grid()
plt.show()
```

```python
# Data Analysis Cell

average_growth = data['GDP % growth'].mean()
print(f"Average GDP Growth (2000-2010): {average_growth:.2f}%")
```

## Summary of Findings
- The analysis covered GDP growth in South Africa from 2000 to 2010.
- The average GDP growth rate during this period was approximately 3.62%.
- Notable trends include  peak growth in 2006, recession in 2009.
- Visualizations indicate.
  
## Insights from GDP % Growth Line Graph

Overall Trend: The line graph indicates a generally upward trend in GDP % growth from 2000 to 2010, highlighting an overall improvement in economic performance.

Year-on-Year Growth: The graph shows fluctuations in GDP growth, beginning with a decline in 2001, which reached a low of -1.7%. After this point, GDP growth began to recover.

Peak Performance: The highest GDP growth rate of 5.6% occurred in 2006, suggesting significant economic expansion during that year.

Stability and Variability: The years following 2001 saw a gradual increase in GDP growth, with notable stability from 2003 to 2005. The average GDP growth rate during this period was around 3.6%, indicating consistent economic performance.

Comparative Growth: The interquartile range shows that 50% of the GDP growth rates fall between 3.0% and 4.95%, emphasizing that most years experienced moderate to strong growth.

Concluding Insight: The overall pattern depicted in the line graph emphasizes a recovering economy, with the most significant growth occurring in 2006. This suggests that economic policies or external factors during that year may have played a crucial role in driving economic expansion.

This analysis highlights the trends in GDP growth in South Africa during the specified period.
